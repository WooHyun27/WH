```python
car_int = car[['Price', 'Age_08_04', 'KM', 'HP', 'Quarterly_Tax', 'Weight', 'Guarantee_Period']]
car_int_corr = car_int.corr(method = 'pearson')
car_int_corr
```

## Price upper outliers analysis

```python
## Price 이상치 값들 구하기
Q3 = car['Price'].quantile(.75)
Q1 = car['Price'].quantile(.25)
IQR = Q3-Q1
outlier_standard = Q3+1.5*IQR

car_outlier = car[car['Price']>=outlier_standard]
car_outlier.info()
```

```python
## outlier model value_counts()
car_outlier['Model'].value_counts()
```
